🧑
## Producto cartesiano
El producto cartesino entre tablas obtiene como resultado todas las filas de una tabla combinada con todas las filas de otra tabla

```python
#@title producto cartesiano de dos tablas
%%sql
select valueA, valueB
from tableA, tableB
order by idA, idB;
```

```python
#@title producto cartesiano de tres tablas
%%sql
select valueA, valueB, valueC
from tableA, tableB, tableC
order by idA, idB, idC;
```

🧑
##inner join
Combina dos tablas pero solo se obtiene las combinaciones que cumplen la condición especificada "ON"

```python
#@title inner join
%%sql
select valueA, valueB
from tableA inner join tableB on (idA = idB);
```

```python
#@title inner join es equivalente a un producto cartesiano filtrado
%%sql
select valueA, valueB
from tableA, tableB
where idA = idB;
```

```python
#@title inner join de tres tablas
%%sql
select valueA, valueB, valueC
from tableA inner join tableB on (idA = idB) inner join tableC on (idB = idC);
```

🧑
##left join
Mismo resultado que inner join pero añade todos las filas de la primera tabla que no se combina con otra fila de la segunda tabla. Las columnas de la segunda tabla tendrán valores NULL

```python
#@title left join
%%sql
select valueA, valueB
from tableA left join tableB on (idA = idB);
```

🧑
#right join
Igual que left join, pero en este caso se incluye las filas de la segunda tabla que no se han combinado con ninguna fila de la otra tabla.

```python
#@title right join
%%sql
select valueA, valueB
from tableA right join tableB on (idA = idB);
```

🧑
#full join
Es la unión de los resultados de left y right join.

Hay sistemas gestores de bases de datos que aceptan la instrucción "full outer join". MySQL no lo soporta, por lo que lo implementamos con el operador "union"

```python
#@title full join - Es la union de left join y rigth join
%%sql
select valueA, valueB
from tableA left join tableB on (idA = idB)

union

select valueA, valueB
from tableA right join tableB on (idA = idB);
```

😸
#join de todos tipos
Aunque lo más habitual es trabajar con join que combina tablas en función de unas columnas que son claves ajenas, puedes hacer cualquier combinación que se te ocurra

```python
%%sql
select t1.valueA, t2.valueA
from tableA t1 inner join tableA t2 on (t1.idA < t2.idA);
```
